```python
import pandas as pd
```

```python
apps_df = pd.read_csv('Play Store Data.csv')
reviews_df = pd.read_csv('User Reviews.csv')
```








### datacleaning

```python
apps_df = apps_df.dropna(subset=['Rating'])
```

```python
for column in apps_df.columns:
    apps_df[column].fillna(apps_df[column].mode()[0], inplace=True)
```

```python
apps_df.drop_duplicates(inplace=True)
```

```python
apps_df = apps_df[apps_df['Rating']<=5]
```

```python
reviews_df.dropna(subset=['Translated_Review'],inplace=True)
```

# Data Transformation

```python
# convert installs coloumn into numeric
apps_df['Installs'] = apps_df['Installs'].str.replace(',','').str.replace('+','').astype(int)


```

```python
# convert price coloumn into float
apps_df['Price'] = apps_df['Price'].str.replace('$','').astype(float)
```

```python
merged_df = pd.merge(apps_df,reviews_df,on='App',how='inner')
```

```python
apps_df['Size'].head(5)
```

```python
merged_df.head()
```

```python
# data transformation
def convert_size(size):
    if 'M' in size:
        return float(size.replace('M',''))
    elif 'K' in size:
        return float(size.replace('K',''))/1024
    else:
        return np.nan

apps_df['Size'] = apps_df['Size'].apply(convert_size)
```

```python
apps_df['Reviews'] = apps_df['Reviews'].astype(int)
```

```python
# log Transformation used in case if data is Right side Skewed to make it log normal distribution
apps_df['LogInstalls'] = np.log(apps_df['Installs'])
apps_df['LogReviews'] = np.log(apps_df['Reviews'])
```

```python
def rating_group(rate):
    if rate >=4:
        return 'Top Rated'
    elif rate >=3:
        return 'Above Average'
    elif rate >=2:
        return 'Average'
    else:
        return 'Below Average'

apps_df['rating_group'] = apps_df['Rating'].apply(rating_group)
    
```

```python
# revenue Coloumn
apps_df['Revenue'] = apps_df['Price']* apps_df['Installs']
```

```python
# sentiment Intensity Analyzer
from nltk.sentiment.vader import SentimentIntensityAnalyzer
```

```python
sia = SentimentIntensityAnalyzer()
```

```python
Reviews = "this app is not so great but can be used for the time being."
Score = sia.polarity_scores(Reviews)
print(Score)
```

```python
Reviews = "this app is very bad. I will not Use it Again!!"
Score = sia.polarity_scores(Reviews)
print(Score)
```

```python
reviews_df['Sentiment_Score'] = reviews_df['Translated_Review'].apply(lambda x: sia.polarity_scores(str(x))['compound'])
```

```python
apps_df['Last Updated'] = pd.to_datetime(apps_df['Last Updated'])
```

```python
apps_df['Year'] = apps_df['Last Updated'].dt.year
```

```python
apps_df['Year'].unique()
```

```python
apps_df['year_month'] = apps_df['Last Updated'].dt.strftime('%B %Y')
```

```python
def save_plot_as_html(fig,filename,insight):
    global plot_containers
    filepath = os.path.join(html_files_path, filename)
    html_content = pio.to_html(fig, full_html= False,include_plotlyjs='incline')
    plot_containers += f"""
    <div class="plot-container" id="{filename}" onclick="openPlot('{filename}')">
        <div class="plot">{html_content}</div>
        <div class="insights">{insight}</div>
    </div>
    """
    fig.write_html(filepath, full_html=False, include_plotlyjs='inline')

    
```

```python
task2_df['year_month'].unique()
```

```python
task2_df['Installs'].unique()
```

```python
task2_grouped = task2_df.groupby(['year_month', 'Category'])['Installs'].sum().reset_index()
```

```python
task2_grouped['year_month'].unique()
```

```python
def month_order(year_month):
    if year_month == 'January 2018':
        return 1
    elif year_month == 'February 2018':
        return 2
    elif year_month == 'March 2018':
        return 3
    elif year_month == 'April 2018':
        return 4
    elif year_month == 'May 2018':
        return 5
    elif year_month == 'June 2018':
        return 6
    elif year_month == 'July 2018':
        return 7
    elif year_month == 'August 2018':
        return 8
    else:
        return 9

task2_grouped['month_order'] = task2_grouped['year_month'].apply(month_order)
```

```python
task2_grouped = task2_grouped.sort_values(by='month_order')

```

```python
task2_grouped['MoM_change'] = task2_grouped.groupby('Category')['Installs'].pct_change() * 100
```

```python
from datetime import datetime
import pytz

#  current time in IST
ist = pytz.timezone('Asia/Kolkata')
current_time = datetime.now(ist).time()

# allowed time range
start_time = datetime.strptime("17:00:00", "%H:%M:%S").time()
end_time = datetime.strptime("19:00:00", "%H:%M:%S").time()
```

```python
import pandas as pd

# Calculate month-over-month percentage change for installs
task2_grouped['Installs_pct_change'] = task2_grouped.groupby('Category')['Installs'].pct_change() * 100

# Identify periods where the percentage change is greater than 20%
highlight_periods = task2_grouped[task2_grouped['Installs_pct_change'] > 20]

# Create the plot
task2_fig = px.line(
    task2_grouped,
    x="year_month",
    y="Installs",
    color="Category",  
    labels={'x': 'time', 'y': 'installs'},
    title='Number of installs over the year',
    color_discrete_sequence=['#AB63FB', '#00FF00', '#FF0000'],  
    width=plot_width,
    height=plot_height
)

# Add shaded regions to highlight the 20% jump periods
if start_time <= current_time <= end_time:
    for idx, row in highlight_periods.iterrows():
        start_date = row['year_month']
        end_date = task2_grouped.iloc[idx + 1]['year_month'] if idx + 1 < len(task2_grouped) else start_date

    task2_fig.add_shape(
        type="rect",  # Rectangle shape
        x0=start_date,  # Start of the period
        x1=end_date,  # End of the period (next month)
        y0=0,  # Start at the bottom of the y-axis
        y1=row['Installs'],  # End at the height of the installs value
        fillcolor="rgba(255, 0, 0, 0.3)",  # Red color with transparency
        line=dict(color="rgba(255, 0, 0, 0.5)", width=1)  # Red outline for the shaded area
    )

    task2_fig.update_layout(
    plot_bgcolor='black',
    paper_bgcolor='black',
    font_color='white',
    title_font={'size': 16},
    xaxis=dict(title_font={'size': 12}),
    yaxis=dict(title_font={'size': 12}),
    margin=dict(l=10, r=10, t=30, b=10)
    )
else:
    print("Graph is hidden. Available only from 5 PM IST to 7 PM IST.")

task2_fig.update_traces(marker=dict(line=dict(color='white',width=1)))
save_plot_as_html(task2_fig,'line chart.html',' time series line chart to show the trend of total installs over time')
```
